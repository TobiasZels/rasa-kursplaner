## Bot Fragen und Antworten


### Daten Erhebung:	

- **Begrüßung:** Willkommen beim Kursplaner, im nachfolgenden werde ich sie bei der Auswahl ihrer Kurse für das nachfolgende Semester unterstützen.
- **Frage nach Studiengang:** Welches Fach studieren sie aktuell als Hauptfach?
- **Frage nach Abschluss:** Studieren sie {{*Hauptfach*}} im Bachelor oder Master Studiengang?
- **Frage nach Nebenfach:** Was ist ihr zweites Hauptfach/ ihre Nebenfächer?
- **Frage nach Semester:** Für welches Semester möchten sie einen Stundenplan erstellen?
- **Frage nach Abgeschlossenen Modulen:** Welche Module haben sie bisher Abgeschlossen?
- **Nutzer kennt seine Module nicht:** Ihre bereits abgeschlossenen Module können sie unter (Flex-Now Link) einsehen.
- **Frage nach Kursen in Angefangenen Modulen:** Haben sie Kurse aus noch nicht abgeschlossenen Modulen bereits besucht?
- **Frage nach geplanter Studiendauer:** Bis zu welchem Semester wollen Sie ihr Studium abschließen, empfohlen wird für *{{Hauptfach}} {{min Semesteranzahl}}* mit einer maximalen Studiendauer von *{{max Semesterzahl}}*.
- **Frage nach geplantem Zeitaufwand:** Wie viel Zeit möchten sie für ihr Studium dieses Semester aufbringen, um das Studium in der gewünschten Zeit abzuschließen werden *{{empfohlene SWS}}* empfohlen.

### Kursauswahl: 



- **Kurs Vorschlagen:** Für das *{{Modul}}* wird der Kurs *{{Kurs}}*, *{{Besonderheiten}}*, angeboten.*{{Zeit slot problem}}* Geben sie an ob sie diesen Kurs belegen wollen oder stellen sie Fragen zu dem Kurs für weitere Informationen.
- **Fragen nach Auswahl des Kurses:** Für das *{{Modul}}* werden *{{Anzahl}}* Kurse, *{{Kurse}}* angeboten.*{{Zeit slot problem}}* Wählen sie ob und welchen Kurs sie belegen wollen oder stellen sie Fragen zu den einzelnen Kursen für weitere Informationen. 
- **Zeit slot problem:** Sollten sie sich für *{{Kurs A}}* entscheiden, so können sie *{{Kurs B}}* aus *{{Modul B}}* nicht mehr wählen, da sich die Zeiten überschneiden.
- **gefundene Informationen:** *{{Kurs}} {{Information}}*
- **keine Informationen:** Diese information liegt für *{{Kurs}}* leider nicht vor.
- **Informationen über mehrere Kurse:** *{{gefundene Informationen}}*, ... und *{{gefundene Informationen}}*.
- **Kurs in Stundenplan aufnehmen:** *{{Kurs}}* wurde in ihren Stundenplan aufgenommen.
- **Verbundene Übungen:** Für *{{Kurs}}* gibt es außerdem die *{{zusammenhängende Übung/Kurse}} {{Informationen}}*. Möchten sie *{{Übung/Kurs}}* belegen?
- **Fragen ob weitere Kurse belegt werden sollen:** Für den Abschluss ihres Studium benötigen sie zusätzlich nur noch *{{LP}}* welche mit dem Kurs *{{Kurs}}* erreicht wäre. Geben sie an ob sie diesen Kurs zusätzlich belegen wollen oder stellen sie Fragen zu dem Kurs für weitere Informationen.
- **Stundenplan ausgeben:** Ihren Stundenplan können sie hier *{{link zu pdf}}* einsehen. Möchten sie den Stundenplan übernehmen oder möchten sie Änderungen an diesem vornehmen.
- **Fragen welcher Kurs gelöscht werden soll:** Welchen Kurs möchten sie aus ihrem Stundenplan entfernen?
- **Kurs kritisch für das Studium:** Sind sie sicher das sie *{{Kurs}}* entfernen wollen, *{{Kurs Relevanz}}*?
- **Kurs hat Übungen:** Möchten sie die Übung zum Kurs *{{Kurs}}* ebenfalls entfernen?
- **Fragen nach Änderung der Filterkriterien:** Für ihre Auswahl konnten keine weiteren Kurse gefunden werden bitte ändern sie ihre Kriterien.
- **Studiumsrelevante Informationen ausgeben:** *{{Studiumsrelevante Informationen}}*
- **Verabschieden:** Ich hoffe ich konnte ihnen weiterhelfen, nicht vergessen ihre gewählten Kurse auf (SPUR Link) zu belegen.
- **Kurs hinzufügen:** *{{Kurs}}* wurde hinzugefügt.
- **Kurs kann nicht hinzugefügt werden:** *{{Kurs}}* kann nicht hinzugefügt werden, da *{{Begründung}}*.

## Kursdaten

Im [Studienportal](https://studierendenportal.uni-regensburg.de/) der Universität Regensburg alle Laschen des gewünschten Studiengangs öffnen und den Seitenquellentext in eine .html datei speichern. (z.B. Medieninformatik.html). Im Anschluss die somit erstellte Datei in den Ordner "Daten" schieben und den Dateinamen unten unter "data" eingeben. Die Vorhandenen Modul- und Kursinformationen werden dann in den nachfolgenden Schritten zu einen JSON object strukturiert und an den vorher angegebenen Speicherort ausgegeben, sodass diese in eine Datenbank eingelesen werden können.

### Imports

`````python
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
`````

### Parse HTML

`````python
# HTML file path of the Course overiew page
data = open("Medieninfo.html","r").read()  
soup = BeautifulSoup(data, 'html.parser')

output = soup.find_all("tr", class_="treeTableCellLevel7")
output2 = soup.find_all("tr", class_="treeTableCellLevel8")
output4 = soup.find_all("tr", class_="treeTableCellLevel6")
print(output2[0])
`````

### Convert to Data Array

`````python
text1 = []
text2 = []
text3 = []

for i in output:
    text1.append(i.text.strip())

for i in output2:
    text2.append(i.text.strip())

for i in output4:
    text3.append(i.text.strip())


course_object = []

for elements in text3:
    _id = elements.split(" ", 1)[0].replace('\n', '')
    _module = elements.split(" ", 2)[1]
    _lp = elements.split(" ", 2)[2].split("-", 1)[0]
    _type = elements.split(" ", 2)[2].split("-", 1)[1].split(" ", 2)[1]
    try:
        _name = elements.split(" ", 2)[2].split("-", 1)[1].split(" ", 2)[2]
    except:
        _name = ""
    _courses = []
    for course in text1:
        if _id in course.split(" ", 1)[0].replace('\n', ''):
            course_id = course.split("\n", 1)[0]
            course_name = course.split(" ", 1)[1].replace("- ", '')
            course_information = []
            for information in text2:
                if course_id in information.split("\n", 1)[0]:
                    group = information.split("\n", 4)[1].replace(' - ', '')
                    try:
                        time = information.split("\n", 4)[2].split(")", 1)[0]
                    except:
                        time = ""

                    

                    try:
                        dozent_replaced = information.split("\n", 4)[2].split(")", 1)[1].replace('Dozent/-in: ', "#")
                        room = dozent_replaced.split('#', 1)[0]
                        teacher = dozent_replaced.split('#', 1)[1].split('Bemerkung', 1)[0]
                        try:
                            additionalInfo = "Bemerkung:" + dozent_replaced.split('#', 1)[1].split('Bemerkung', 1)[1]
                        except:
                            additionalInfo = ""
                    except:
                        room = ""
                        teacher = ""
                        additionalInfo = ""
                    course_information.append({"group": group,
                                    "time": time,
                                    "room": room,
                                    "additionalInfo": additionalInfo,
                                    "teacher": teacher})


            _courses.append({"id": course_id,
                            "name": course_name,
                            "information": course_information})


    course_object.append({"id": _id,
                        "module": _module,
                        "lp": _lp,
                        "type": _type,
                        "name": _name,
                        "courses": _courses})

text3_split = text3[0].split(" ", 2)

pprint(course_object)
`````
